# Importing required libraries

```python
import pandas as pd
import numpy as np
import getpass
from sqlalchemy import create_engine
import pymysql
```

# To connect and extract tables from SQL database.

```python
driver = 'mysql+pymysql'
user = 'root'
password = getpass.getpass(prompt='Your password:'"")
ip = '127.0.0.1'
```

```python
connection_string = f'{driver}://{user}:{password}@{ip}'
db_connection = create_engine(connection_string)
```

```python
pokemon = pd.read_sql_query('SELECT * FROM pokemon.pokemon_stats;', db_connection)
pokemon.head()
```

```python
query = "SELECT * FROM sakila.actor"
df = pd.read_sql_query(query, db_connection)
df.head()
```

```python
query = 'SELECT * FROM orders.orders;'
orders = pd.read_sql_query(query, db_connection)
orders.drop(columns='index', inplace=True)
orders.head()
```

```python
query = 'SELECT * FROM publications.authors;'
publications = pd.read_sql_query(query, db_connection)
publications.head()
```

```python
query = """select track_name, rating_count_tot
from appleStore.data
where rating_count_tot != 0
order by rating_count_tot desc
limit 10;
"""

apple_store_most_rated_apps = pd.read_sql(query, db_connection)

apple_store_most_rated_apps
```

# Create a database in Mysql and save a table from Pandas into that database.

```python
import mysql.connector

driver = 'mysql+pymysql'
user = 'root'
password = getpass.getpass(prompt='Your password:')
ip = '127.0.0.1'
database = 'testing_123'
```

```python
mydb = mysql.connector.connect(host=ip,  user=user, password=password)
mycursor = mydb.cursor()
mycursor.execute(f"CREATE DATABASE {database};")
```

```python
connection_string = f'{driver}://{user}:{password}@{ip}/{database}'
db_connection = create_engine(connection_string)
```

```python
apple_store_most_rated_apps.to_sql('table_1', db_connection)
```
